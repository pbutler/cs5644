```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 50)
y1 = np.sin(x)
y2 = np.cos(x)
yrandom = np.random.random(50)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
ax1 = plt.subplot(2, 1, 1)
ax1.plot(x, y1)
ax2 = plt.subplot(2, 1, 2)
ax2.plot(x, y2)
```

```python
for y in (y1, y2):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(1,1,1)
    ax.plot(x, y)
```

```python
plt.hist([y1, y2, yrandom])
```

```python
plt.scatter(y1, y2, s=yrandom*100)
```

```python
plt.scatter(y1, y2, c=yrandom)
```
